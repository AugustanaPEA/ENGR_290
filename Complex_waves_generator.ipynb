```python
import numpy as np
import pandas as pd

t_array = np.linspace(0,10,1001)

dc = 15.0
freq1 = 1.0; amp1 = 4.0; ph1 = 0.0
freq2 = 3.0; amp2 = 1.0; ph2 = np.pi/2
freq3 = 20.0; amp3 = 0.4; ph3 = 0.0

dc_val = t_array*0.0 + dc
val1 = amp1*np.sin(t_array*freq1+ph1)
val2 = amp2*np.sin(t_array*freq2+ph2)
val3 = amp3*np.sin(t_array*freq3+ph3)

val_sum = 0.0
val_sum = dc + val1 + val2 + val3
data = pd.Series(data=val_sum, index=t_array)
data.plot(xlabel = 'time(s)', ylabel='amplitude', ylim = [10,22],
          figsize=[8.0,4.0]);
```
